# Introduction to Linear Algebra

Linear algebra is a branch of mathematics that focuses on linear equations, vectors, and matrices. It is a fundamental building block for many areas of science and engineering, including machine learning and computer vision.

## 1. Vectors

A vector is an ordered list of numbers. It can be thought of as a point in space or an arrow with a specific magnitude and direction.

A vector $v$ in $\mathbb{R}^n$ is denoted as:
$$ v = \begin{bmatrix} v_1 \\ v_2 \\ \vdots \\ v_n \end{bmatrix} $$

Let's start by importing NumPy, the most popular library for numerical computing in Python.

```python
import numpy as np

# Creating a vector as a NumPy array
v = np.array([1, 2, 3])
print("Vector v:", v)
print("Shape of v:", v.shape)
```

## 2. Matrices

A matrix is a rectangular grid of numbers. An $m \times n$ matrix $A$ has $m$ rows and $n$ columns.

$$ A = \begin{bmatrix} a_{11} & a_{12} & \dots & a_{1n} \\ a_{21} & a_{22} & \dots & a_{2n} \\ \vdots & \vdots & \ddots & \vdots \\ a_{m1} & a_{m2} & \dots & a_{mn} \end{bmatrix} $$

```python
# Creating a 2x3 matrix
A = np.array([[1, 2, 3], 
              [4, 5, 6]])

print("Matrix A:\n", A)
print("Shape of A:", A.shape)
```

## 3. Basic Operations

### Scalar Multiplication
Multiplying a vector or matrix by a single number (a scalar) multiplies each element by that number.

### Matrix Multiplication
Matrix multiplication is one of the most important operations in linear algebra. Given a matrix $A$ of size $m \times n$ and $B$ of size $n \times p$, the product $C = AB$ is a matrix of size $m \times p$.

```python
# Scalar Multiplication
v_scaled = 2 * v
print("2 * v:", v_scaled)

# Matrix-Vector Multiplication
# Note: A is 2x3 and v is 3x1 (interpreted by @)
result = A @ v
print("A @ v:", result)
```

## What is a Jupyter Notebook?

A Jupyter Notebook is an interactive document that contains both code and rich text elements (paragraphs, equations, figures, links, etc.). 

Notebooks are made up of **cells**. There are two main types of cells:
1. **Markdown Cells**: These cells contain text, like this one! You can use Markdown to format your text.
2. **Code Cells**: These cells contain code that you can execute.

## 1. Markdown Basics

You can use Markdown to create:
- **Bold text**
- *Italic text*
- [Links to websites](https://jupyter.org)
- Lists (like this one!)
- Even math equations using LaTeX: $E = mc^2$

## 7. Data Visualization

Now that we've installed the libraries, let's create a simple plot using `numpy` and `matplotlib`.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create the plot
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='Sine Wave', color='blue')
plt.title('Simple Plot in Jupyter')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.grid(True)
plt.show()
```

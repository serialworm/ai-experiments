## What is a Jupyter Notebook?

A Jupyter Notebook is an interactive document that contains both code and rich text elements (paragraphs, equations, figures, links, etc.). 

Notebooks are made up of **cells**. There are two main types of cells:
1. **Markdown Cells**: These cells contain text, like this one! You can use Markdown to format your text.
2. **Code Cells**: These cells contain code that you can execute.

## 1. Markdown Basics

You can use Markdown to create:
- **Bold text**
- *Italic text*
- [Links to websites](https://jupyter.org)
- Lists (like this one!)
- Even math equations using LaTeX: $E = mc^2$

## 2. Running Code

The cell below is a **Code Cell**. You can run it by selecting it and pressing `Shift + Enter`.

```python
message = "Hello, Jupyter!"
print(message)
```

## 3. Variable Persistence

One of the coolest features of Jupyter is that variables you define in one cell are available in other cells. 

Run the cell below, which uses the `message` variable we defined in the previous cell.

```python
print(f"The message was: {message}")
number = 10
print(f"New variable 'number' is: {number}")
```

## 4. Automatic Output

In Jupyter, the result of the last line in a code cell is automatically displayed, even if you don't use `print()`.

```python
a = 5
b = 7
a + b
```

## 5. Rich Outputs

Jupyter can display more than just text. If you have libraries like `matplotlib` or `pandas` installed, you can show plots and tables directly in the notebook.

```python
import math
print(f"The square root of 144 is {math.sqrt(144)}")
```
